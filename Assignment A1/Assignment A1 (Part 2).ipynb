```
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab.patches import cv2
#cv = open source library with all things related to computer vision
```

```
image = cv2.imread('/content/drive/MyDrive/CAS CS 585/Assignment A1/input_image.jpg')
```

```
#print(type(image))
#print(image.shape)

#first value = number of rows of pixels in the image
#second value = number of columns of pixels in the image
#third value = blue, green, red (ocv order)

cv2_imshow(image)
```

```
#grayscale image of my face:
def grayscale (image):
  for row in range (image.shape[0]):
    for col in range (image.shape[1]):
      image[row,col,:] = np.average(image[row,col,:])

gray_image = image.copy()
grayscale(gray_image)
cv2_imshow(gray_image)
```

```
#inverted image of my face:
def inverted (image):
  for row in range (image.shape[0]):
    for col in range (image.shape[1]):
      image[row, col, 0] = 255 - image[row, col, 0]
      image[row, col, 1] = 255 - image[row, col, 1]
      image[row, col, 2] = 255 - image[row, col, 2]
inverted_image = image.copy()
inverted(inverted_image)
cv2_imshow(inverted_image)
```
